# Kompendium yeast fedbatch

```python
run -i BPL_YEAST_AIR_Fedbatch_DOcontrol_explore.py
```

```python
plt.rcParams['figure.figsize'] = [30/2.54, 25/2.54]
```

```python
newplot(title='Yeast fedbatch cultivation',plotType='Textbook_2_half')

# Culture parameters
par(qO2lim=0.0069) 

# Process initial conditions
init(V_0=4.5, VG_0=4.5*30, VX_0=4.5*2, VE_0=4.5*0) 

# Feed profile
par(t_start=9, F_start=0.044, mu_feed=0.20, F_max=0.18)  

# DO-control parameters
par(samplePeriod=1/60, K=10, Ti=0.5, I_0=500)            

simu(20)
```

```python
newplot(title='Yeast fedbatch cultivation',plotType='Textbook_2')
show()
```
