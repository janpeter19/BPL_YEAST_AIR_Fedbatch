# Notes YEAST AIR Fedbatch DO-control

Here we run application YEAST AIR with fedbatch operation and DO-control. First a conservative exponentail feed profile with margins to ethanol production. In the next diagram manual adjustments of feed rate is done every 2.5 hours and the culture swtiches bewteen etanol production and consumption. More than double amount of cells are produced. 

At the end we touch questions around numerical properties of the simulation-alogorithm.

```python
# Culture
par(Ks=1E-3)

# Process initialization
par(V_tot=10)
init(V_0=4.5, VX_0=4.5*1.0, VG_0=4.5*5.0, VE_0=0) 

# Feeding
par(G_in=500, t_start=6, F_start=0.011, F_max=0.07, mu_feed=0.20)    

# Aeration
par(airFlow_setpoint=120, alpha_O2=1, alpha_E=0)

# Simulation
newplot(); ax3.set_ylim([0,110]); ax6.set_ylim([400,1000])
simu()
```

Note that in the second example the ethanol response is slightly different from when the application YEAST without gas phase. This is due to the fact that a small amount of ethanol evaporates to th gas phase and out from the reactor. The parameter alpha_E describe the evatporation and when set to zero the same patterna as without gas phase is obtained.

Note that some of the DO-N transients are difficult to understand
* At 12.5 on - difficult
* At 15.0 on - likely similar as the one before
* At 17.5 on - short starvation just before and leads to drop in N

Possible with some numerical difficulties for the first two and would help change numerical simulation settings?
